```python
#label encoder. mengubah status sehat menjadi 1, status tidak sehat menjadi 0.
encoder = LabelBinarizer()
encoder.fit(mydata_train["Kelas Status Pulang Peserta"])

mydata_train["Kelas Status Pulang Peserta"] = encoder.transform(mydata_train["Kelas Status Pulang Peserta"])
```

```python
#split data train set
x = mydata_train.iloc[:,0:18].values
y = mydata_train.iloc[:,18:22].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_val, y_train, y_val = train_test_split(x, y, test_size=0.2, random_state = 0)
```

```python
import warnings

warnings.filterwarnings('ignore')
from sklearn.ensemble import RandomForestClassifier
rfc_100 = RandomForestClassifier(n_estimators=100, random_state=0)



# fit the model to the training set

rfc_100.fit(x_train, y_train)



# Predict on the test set results

y_pred_100 = rfc_100.predict(x_val)



# Check accuracy score 

print('Model accuracy score with 100 decision-trees : {0:0.4f}'. format(accuracy_score(y_val, y_pred_100)))
```

```python
# Print the Confusion Matrix and slice it into four pieces

from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_val, y_pred_100)

print('Confusion matrix\n\n', cm)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_val, y_pred_100))
```
